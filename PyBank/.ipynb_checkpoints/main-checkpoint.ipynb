```python
# import pandas
import pandas as pd
from pathlib import Path

# set the file path, read in csv from filepath
file_path = Path("../../Desktop/nu-chi-fin-pt-07-2021-u-c-master/02-Homework/02-Python/Instructions/PyBank/Resources/budget_data.csv")
budget_df = pd.read_csv(file_path)

#set variables
total_months=[]
total_net_profit=[]
mean_profit_change=[]
max_increase_value=[]
max_decrease_value=[]
max_increase_month=[]
max_decrease_month=[]

#shape will give us how many rows(months) correspond with the profit/losses inputs
total_months = budget_df.shape[0]

# sum up the profit/loss column values to get net profit
column_name = "Profit/Losses"
total_net_profit = budget_df[column_name].sum()

# Set the date as the index to the DataFrame
budget_df = budget_df.set_index("Date")
budget_df.head()

#calculate difference from month to month 
monthly_profit_change = budget_df.diff()

#find average of the changes in Profit/Losses over the entire period
mean_profit_change = monthly_profit_change["Profit/Losses"].mean()

#find greatest increase in profits
max_increase_value = monthly_profit_change["Profit/Losses"].max()

#find greatest decrease in profits
max_decrease_value = monthly_profit_change["Profit/Losses"].min()

#find the month that corresponds to the max increase in monthly profit change
max_increase_month = monthly_profit_change.loc[monthly_profit_change['Profit/Losses'] == max_increase_value ]

#find the month that corresponds to the max decrease in monthly profit change
max_decrease_month = monthly_profit_change.loc[monthly_profit_change['Profit/Losses'] == max_decrease_value ]


#print values in statements
print("Financial Analysis")
print("-------------------------")
print(f"Total Months: {total_months}")
print(f"Total Net Profit: ${total_net_profit}")
print(f"Average  Change: ${round(mean_profit_change,2)}")
print(f"Greatest Increase in Profits: ${round(max_increase_value)}")
print(f"Greatest Decrease in Profits: ${round(max_decrease_value)}")


#write print statements, set output path for .txt file
results_output = Path("PyBank_Results.txt")

with open(results_output,"w") as file:
    file.write(f"Financial Analysis")
    file.write("\n")
    file.write(f"-------------------------")
    file.write("\n")
    file.write(f"Total Months: {total_months}")
    file.write("\n")
    file.write(f"Total Net Profit: ${total_net_profit}")
    file.write("\n")
    file.write(f"Average  Change: ${round(mean_profit_change,2)}")
    file.write("\n")
    file.write(f"Greatest Increase in Profits: ${round(max_increase_value)}")
    file.write("\n")
    file.write(f"Greatest Decrease in Profits: ${round(max_decrease_value)}")



```
